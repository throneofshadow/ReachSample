### Next, let's create and visualize robot commands for a single spatial dimension of task. We will select varying positions in the $ \theta $ horizontal axis command position that either have randomness within the task or conserve the entropy of task positions. Command positions can additionally be varied over a selected number of trials, length of vector, and radius from arena entrance. Let's visualize two sets of 1-D spaces. One set will contain constant points and define the range of our experiments. The other set will randomly sample between non-control points. 

```python
RS_.create_theta_workspace(1.5, 2, 50, 9, visualize = True, export=True, export_filename='theta_commands.csv')
```

```python
RS_.create_theta_workspace(1.5, 2, 100, 9, sample=True, visualize = True, export=True, export_filename='theta_commands_random.csv')
```

```python
RS_.create_phi_workspace(1.5, 2, 100, 9, sample=True, visualize = True, export=True, export_filename='phi_commands_random.csv')
```

## Let's look at some different workspaces now. Below is an example of a 2-D workspace that is completely symmetric, as well as one that has slightly smaller variation in the z-plane.

```python
RS_.create_2d_workspace(1.5, 1.5, 2, 150, 9, sample=True, visualize = True, export=True, export_filename='2d_commands_random.csv')
```

## To add depth to our task positions without biasing comparisons of different points w.r.t. distance, we stride our 2-D workspaces by a constant both in the +x and -x direction. This allows us to fully sample the 3-D workspace. Additionally, we also can sample to fixed, more symmetric locations. This is useful if a researcher needs more direct comparison data.

```python
RS_.create_3d_workspace(0.25, 1.5, 1.5, 2, 150, 9, sample=True, visualize = True, export=True, export_filename='3d_commands.csv')
```

```python
RS_.create_3d_workspace(0.5, 0.4, 1, 2, 1, 9, sample=True, visualize = True)
```
